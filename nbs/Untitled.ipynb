```python
from bs4 import BeautifulSoup
import requests
from bs4 import BeautifulSoup
```

```python
def word_translation(dataframe):
    for i in range(len(dataframe)):
        wo_word = str(dataframe.iloc[i]['title'])
    if len(wo_word.split(" "))>1:
        wo_word = '%20'.join(wo_word.split(" "))
    page_link = 'https://en.glosbe.com/wo/en/' + wo_word
    page_response = requests.get(page_link, timeout=5)
    soup = BeautifulSoup(page_response.content, 'html.parser')
    alert = [i.text for i in soup.findAll('p',attrs={"class":"alert alert-info"})]
    if len(alert) != 0:
        dataframe['translated'][i] = '--unk--'
    else:
        dataframe['translated'][i] = [i.text for i in soup.find_all('strong')[1:]]
        dataframe['explanation'][i] = [i.text for i in soup.findAll('div',attrs={"class":"phraseMeaning show-user-name-listener"})]
    
```
